```python
# We do not need the top model so we drop it
df1.drop(['BaseModel', 'TopModel'], axis=1, inplace=True)
```

```python
# Drop row 10 as range is null
df1.drop(10, inplace=True)
```

```python
df1.drop(['index'], axis=1, inplace=True)
```

```python
#Preprocessing Range column
for i in range(11):
    df1['Range'][i] = df1['Range'][i].split(' ')[0]
```

```python
df1
```

```python
# Preprocessing Capacity coulmn
for i in range(11):
    df1['Capacity'][i] = df1['Capacity'][i].split(' ')[0]
```

```python
df1
```

```python
# Dropiing Vehicle Type as we are working on EVs only so no need to specify it
df1.drop(['VehicleType'], axis=1, inplace=True)
```

```python
df1
```

```python
df = df1.copy()
```

```python
df
```

```python
df['BootSpace'].isnull()[4]
```

```python
# Preprocessing BootSpace Column
for i in range(11):
    if df['BootSpace'].isnull()[i] == False:
        df['BootSpace'][i] = df['BootSpace'][i].split(' ')[0]
```

```python
df
```

```python
df['BootSpace'] = pd.to_numeric(df['BootSpace'])
```

```python
df['BootSpace']
```

```python
# Calculating mean BootSpace to replace null
mean_value = df['BootSpace'].mean()
mean_value
```

```python
df['BootSpace'].fillna(value=mean_value, inplace=True)
```

```python
df
```

```python
df['PriceRange'][6].split(' ')[-1]
```

```python
for i in range(11):
    if df['PriceRange'][i].split(' ')[-1] == 'L':
        if len(df['PriceRange'][i].split(' ')) == 4:
            avg = ((float(df['PriceRange'][i].split(' ')[0]) + float(df['PriceRange'][i].split(' ')[2]))/2)*100000
            df['PriceRange'][i] = avg
        else:
            df['PriceRange'][i] = float(df['PriceRange'][i].split(' ')[0])*100000
    else:
        if len(df['PriceRange'][i].split(' ')) == 4:
            avg = ((float(df['PriceRange'][i].split(' ')[0]) + float(df['PriceRange'][i].split(' ')[2]))/2)*10000000
            df['PriceRange'][i] = avg
        else:
            df['PriceRange'][i] = float(df['PriceRange'][i].split(' ')[0])*10000000
```

```python
df
```

```python
# Dropping Transmission as it will not have any effect on model as all value are same
df.drop(['Transmission'], axis=1, inplace=True)
```

```python
df
```

```python
df['Range'] = pd.to_numeric(df['Range'])
df['Capacity'] = pd.to_numeric(df['Capacity'])
```

```python
# Saving Preprocessed Data as Excel Workbook
df.to_excel('EVIndia-Preprocessed.xlsx', index=False)
```
